```python
def initialize_centroids(X,k):
    m,n=X.shape
    centroids=np.zeros((k,n))
    q=np.random.randint(m,size=(1,k))
    for i in range(k):
        centroids[i]=X.ix[q[0][i]]
    return centroids
'''a=np.array([[1],[2],[3]])
print(a.shape)
b=np.amax(a)
c=np.argmax(a)
print(b,c)'''

#print(initial_centroids)
```

```python
'''a=np.array([1,2,3])
b=pd.DataFrame(a)
c=b.values
print(type(c))
print(b)'''
def findclosestcentroids(X,centroids):
    X_new=X.values
    m,n=X.shape
    k=centroids.shape[0]
    idx=np.zeros((m,1))
    y=np.zeros((k,1))
    for i in range(m):
        for j in range(k):
            y[j][0]=np.linalg.norm(np.subtract(X_new[i][:],centroids[j][:]))
        b=np.argmin(y)
        idx[i][0]=b
    return idx

#print(idx)
        
        
    
```

```python
def computecentroids(X,idx,k):
    X_new=X.values
    m,n=X.shape
    centroids=np.zeros((k,n))
    mean=np.zeros((k,n))
    for i in range(k):
        c=0
        for j in range(m):
            if (idx[j][0]==i):
                mean[i][:]=np.add(mean[i][:],X_new[j][:])
                c+=1
        centroids[i][:]=np.divide(mean[i][:],c)
    return centroids

#print(centroids)
```

```python
def runkmeans(X,initial_centroids,max_iters):
    X_new=X.values
    m,n=X.shape
    k=initial_centroids.shape[0]
    centroids=initial_centroids
    for i in range(max_iters):
        idx=findclosestcentroids(X,centroids)
        centroids=computecentroids(X,idx,k)
    idx=findclosestcentroids(X,centroids)
    return centroids,idx

#print(centroids)
#print(idx)
```

```python
def elbowmethod(X,k):
    K=np.zeros((1,k))
    m,n=X.shape
    X_new=X.values
    for i in range(k):
        initial_centroids=initialize_centroids(X,i+1)
        centroids,idx=runkmeans(X,initial_centroids,100)
        for j in range(m):
            q=np.subtract(X_new[j][:],centroids[int(idx[j][0])][:])
            v=np.power(q,2)
            K[0,i]=K[0,i]+np.sum(v)
        K[0,i]=K[0,i]/m
    plt.plot(range(1,k+1),K.T)
    plt.xlabel("no.of clusters")
    plt.ylabel("cost")
    plt.show()
elbowmethod(iris_dataframe,10)
```
